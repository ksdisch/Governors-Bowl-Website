```python
total_reg_season_matchups = 0
for year in range(2016, 2022):
    league_id = 439215
    year = year
    swid = '{DC053E2F-AEFD-4F3F-853E-2FAEFD7F3F35}'
    s2 = 'AEBlfvLxcuJ4XnkE79EumyvuDZ8Mii6m%2B6aDD461H3tVa6rVqzan30psYWcMotL%2FnIP6RR4S6ESi4ueZiYXzXkpvBp3zDaIPKQOU9SOXIoU1OzFPAjkWEAVSbFMFqIx6%2FKaa9sH3oe%2BYYyPckt1pULwyi37N5Yzrw39GFh4weR5Kz7CAh%2BOXayJ%2BjMENuDA%2F0J5hXQoZ9pWKXI3NpbEpiabTvPsM8JRL1jLSYGDNAH764wVxUkYmLT%2F3UgQ%2FV47f6rM1aEm9KVdO%2BV%2FLilrNCivv2WGQxFt1gaaJT85oPcMtyw%3D%3D'
    league = League(league_id, year, s2, swid)
    teams = league.teams
    reg_season_length = len(teams[0].schedule) - 3
    print(reg_season_length)
    total_reg_season_matchups += reg_season_length
    # should make a function
    # loop through teams to get all stats for particular season and add them to corresponding values in all_time_team_stats
    for team in teams:
        owner = team.owner.lower()
        #get the stats for the team for the season (reg season only)
        season_wins = team.wins
        season_losses = team.losses
        season_ties = team.ties
        season_pf = team.points_for
        season_pa = team.points_against
        season_acquisitions = team.acquisitions
        season_budget_spent = team.acquisition_budget_spent
        season_trades = team.trades
        #   season_longest_win_strk = team (figure this out later)
        playoff_seed = team.standing
        final_rank = team.final_standing
        projected_rank = team.draft_projected_rank

        # add those to the totals in all_time_team_stats
        try:
            all_time_team_stats[owner]["total_wins"] += season_wins
            all_time_team_stats[owner]["total_losses"] += season_losses
            all_time_team_stats[owner]["ties"] += season_ties
            all_time_team_stats[owner]["total_points_for"] += season_pf
            all_time_team_stats[owner]["total_points_against"] += season_pa
            all_time_team_stats[owner]["total_acquisitions"] += season_acquisitions
            all_time_team_stats[owner]["budget_spent"] += season_budget_spent
            all_time_team_stats[owner]["total_trades"] += season_trades
            all_time_team_stats[owner]["average_playoff_seed"] += playoff_seed
            all_time_team_stats[owner]["average_final_rank"] += final_rank
        except KeyError:
            owner = "brian wilson"
            all_time_team_stats[owner]["total_wins"] += season_wins
            all_time_team_stats[owner]["total_losses"] += season_losses
            all_time_team_stats[owner]["ties"] += season_ties
            all_time_team_stats[owner]["total_points_for"] += season_pf
            all_time_team_stats[owner]["total_points_against"] += season_pa
            all_time_team_stats[owner]["total_acquisitions"] += season_acquisitions
            all_time_team_stats[owner]["budget_spent"] += season_budget_spent
            all_time_team_stats[owner]["total_trades"] += season_trades
            all_time_team_stats[owner]["average_playoff_seed"] += playoff_seed
            all_time_team_stats[owner]["average_final_rank"] += final_rank
                
# calculate averages outside yearly for loop
for team in teams:
    owner = team.owner.lower()
    try:
        all_time_team_stats[owner]["avg_wins"] += all_time_team_stats[owner]["total_wins"] / len(range(2016, 2022))
        all_time_team_stats[owner]["avg_losses"] += all_time_team_stats[owner]["total_losses"] / len(range(2016, 2022))
        all_time_team_stats[owner]["yearly_avg_points_for"] += all_time_team_stats[owner]["total_points_for"] / len(range(2016, 2022))
        all_time_team_stats[owner]["weekly_avg_points_for"] = all_time_team_stats[owner]["total_points_for"] / total_reg_season_matchups
        all_time_team_stats[owner]["yearly_avg_points_against"] += all_time_team_stats[owner]["total_points_against"] / len(range(2016, 2022))
        all_time_team_stats[owner]["weekly_avg_points_against"] = all_time_team_stats[owner]["total_points_against"] / total_reg_season_matchups
        all_time_team_stats[owner]["average_playoff_seed"] = all_time_team_stats[owner]["average_playoff_seed"] / len(range(2016, 2022))
        all_time_team_stats[owner]["average_final_rank"] = all_time_team_stats[owner]["average_final_rank"] / len(range(2016, 2022))
    except KeyError:
        owner = "brian wilson"
        all_time_team_stats[owner]["avg_wins"] += all_time_team_stats[owner]["total_wins"] / len(range(2016, 2022))
        all_time_team_stats[owner]["avg_losses"] += all_time_team_stats[owner]["total_losses"] / len(range(2016, 2022))
        all_time_team_stats[owner]["yearly_avg_points_for"] += all_time_team_stats[owner]["total_points_for"] / len(range(2016, 2022))
        all_time_team_stats[owner]["weekly_avg_points_for"] = all_time_team_stats[owner]["total_points_for"] / total_reg_season_matchups
        all_time_team_stats[owner]["yearly_avg_points_against"] += all_time_team_stats[owner]["total_points_against"] / len(range(2016, 2022))
        all_time_team_stats[owner]["weekly_avg_points_against"] = all_time_team_stats[owner]["total_points_against"] / total_reg_season_matchups
        all_time_team_stats[owner]["average_playoff_seed"] = all_time_team_stats[owner]["average_playoff_seed"] / len(range(2016, 2022))
        all_time_team_stats[owner]["average_final_rank"] = all_time_team_stats[owner]["average_final_rank"] / len(range(2016, 2022))
# all_time_team_stats
print(total_reg_season_matchups)
```
