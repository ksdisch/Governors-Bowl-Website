# Matchup Stats

```python
current_year = date.today().year
print(current_year)
league_id = 439215
year = 2021
swid = '{DC053E2F-AEFD-4F3F-853E-2FAEFD7F3F35}'
s2 = 'AEBlfvLxcuJ4XnkE79EumyvuDZ8Mii6m%2B6aDD461H3tVa6rVqzan30psYWcMotL%2FnIP6RR4S6ESi4ueZiYXzXkpvBp3zDaIPKQOU9SOXIoU1OzFPAjkWEAVSbFMFqIx6%2FKaa9sH3oe%2BYYyPckt1pULwyi37N5Yzrw39GFh4weR5Kz7CAh%2BOXayJ%2BjMENuDA%2F0J5hXQoZ9pWKXI3NpbEpiabTvPsM8JRL1jLSYGDNAH764wVxUkYmLT%2F3UgQ%2FV47f6rM1aEm9KVdO%2BV%2FLilrNCivv2WGQxFt1gaaJT85oPcMtyw%3D%3D'
league = League(league_id, year, s2, swid)
```

```python
weekly_box_scores = league.box_scores(week=17)
weekly_box_scores
```

```python
matchups_list = []
weekly_scores = {}
```

```python
for matchup in weekly_box_scores:
    home_team = matchup.home_team
    home_score = matchup.home_score
    away_team = matchup.away_team
    away_score = matchup.away_score
    if home_score > away_score:
        winner = home_team
        winner_score = home_score
        loser = away_team
        loser_score = away_score
    elif away_score > home_score:
        winner = away_team
        winner_score = away_score
        loser = home_team
        loser_score = home_score
    matchups_list.append({
        "winner": str(winner)[5: -1],
        "winner_score": winner_score,
        "loser": str(loser)[5: -1],
        "loser_score": loser_score,
        "victory_margin": round(winner_score - loser_score, 2)
    })
    weekly_scores[str(home_team)[5:-1]] = home_score
    weekly_scores[str(away_team)[5:-1]] = away_score
```

# Superlatives
#### Highest scorer, lowest scorer, biggest blowout, closest matchup, 
#### top performer at each position (QB, RB, WR, TE, D/ST, K), 
#### biggest bust at each position (QB, RB, WR, TE, D/ST, K)


```python
# High scorer and low scorer
high_score = max(weekly_scores.values())
high_scoring_team = max(weekly_scores, key=weekly_scores.get)
low_score = min(weekly_scores.values())
low_scoring_team = min(weekly_scores, key=weekly_scores.get)
print(high_scoring_team)
print(high_score)
print(low_scoring_team)
print(low_score)
league.top_scorer()
```

```python
#biggest blowout and closest matchup
victory_margins_list = []
for matchup in matchups_list:
    victory_margins_list.append(matchup["victory_margin"])

max_victory_margin = max(victory_margins_list)
min_victory_margin = min(victory_margins_list)
for matchup in matchups_list:
    if matchup["victory_margin"] == max_victory_margin:
        biggest_blowout_matchup = matchup
    elif matchup["victory_margin"] == min_victory_margin:
        narrowest_matchup = matchup


print(biggest_blowout_matchup)
print(narrowest_matchup)
```
